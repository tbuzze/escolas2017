```python
#contando a quantidade de escolas por município
db['NO_MUNICIPIO'].value_counts()
```

```python
#carregando a base do PIB
db_pib = pd.read_csv("PIB dos Municípios - base de dados 2010-2015.csv", sep=';', engine='python')
```

```python
#printando a head da base
db_pib.head()
```

```python
#filtrando a base pelo ano de 2015
db_pib_2015 = db_pib[db_pib['Ano']==2015]
```

```python
#renomeando a coluna 16 para PIB (R$ 1000)
db_pib_2015.columns.values[16] = 'PIB (R$1000)'
```

```python
#printando o head da base
db_pib_2015.head()
```

```python
#copiando da base PIB apenas as colunas neceessárias
db_pib_2015_filt = db_pib_2015.iloc[:,[3,4,16]]
```

```python
#printando a base
db_pib_2015_filt.head()
```

```python
#renomeando a coluna 0
db_pib_2015_filt.columns.values[0] = 'CO_MUNICIPIO'

#criando uma base nova de PIB apenas com os campos necessários
db2 = db_pib_2015_filt.copy()
```

```python
#printando o head da base
db2.head()
```

```python
#fazendo um join left entre a base das escolas e a base do PIB
result = pd.merge(db, db2, how='left', on='CO_MUNICIPIO')
```

```python
#checando o tamanho da nova base
result.shape
```

```python
#printando o head da nova base
result.head(10)
```

```python
#agrupando a nova base pelo Código do Município
group = result.groupby(['CO_MUNICIPIO', 'PIB (R$1000)']).agg(['count', 'max'])
group
```

```python
#contando a quantidade de escolas por código de município
result['CO_MUNICIPIO'].value_counts()
```

```python
#copiando para mun o resultado do grpuby do código de município
mun = result.groupby('CO_MUNICIPIO') 
MUN_count = mun['CO_MUNICIPIO'].count()

#copiando para PIB o resultado do grpuby do código de município
PIB = mun['PIB (R$1000)'].max()
PIB_num = PIB.values 
```

```python
#importando a biblioteca numpy
import numpy as np
```

```python
#transformando as bases em numpy array
PIB_array = np.array(PIB_num, dtype=np.float)
MUN_array = np.array(MUN_count, dtype=np.float)
```

```python
#calculando a taxa do PIB por quantidade de escolas no município
rel_pib_mun = PIB_array/MUN_array
rel_pib_mun = pd.DataFrame(rel_pib_mun)
```

```python
#printando o resultado
rel_pib_mun
```

```python
#transformando a base em uma numpy array
rel_pib_mun_array = np.array(rel_pib_mun)
rel_pib_mun_array

#criando uma base com valores únicos de código do município
unique = np.array(result['CO_MUNICIPIO'].unique())
#unique
```

```python
#criando um dataframe com o PIB e o código da cidade
df = pd.DataFrame(data=np.column_stack((rel_pib_mun_array,unique,PIB_array, MUN_array)),columns=['PIB','Cidade','PIB_Geral', 'Escolas'])
```

```python
#printando a head da base
df.head()
```

```python
#left join para trazer o nome das cidades pelo código
db_cidade = db.iloc[:,4:6]

#renomeando a coluna 0
db_cidade.columns.values[1] = 'Cidade'

#criando uma base nova de PIB apenas com os campos necessários
db_cidadec = db_cidade.copy()
db_cidadec.head()
```

```python
#trazendo o nome das cidade para o dataset
merge_cidade = pd.merge(df, db_cidadec, how='left', on='Cidade')
merge_cidade.head()
```

```python
#excluindo os valores repetidos da coluna cidade
merge_cnod = merge_cidade.drop_duplicates(subset='Cidade')
```

```python
#ordenando o data frame pelos valores do PIB
merge_cnod = merge_cnod.sort_values(by='PIB', ascending=False)
merge_cnod2 = merge_cnod.reset_index(drop=True)
```

```python
#printando os valores do head
merge_cnod2.head()

#retornando a posição de são paulo
merge_cnod2[ merge_cnod2['NO_MUNICIPIO'] == 'São Paulo']
```

```python
#copiando as 20 primeiras linhas do dataframe
grafico_pib = merge_cnod2.iloc[0:20,]
```

```python
#printando as linhas
grafico_pib.head(20)
```

```python
#plotando o gráfico das 20 primeiras cidades com o Maior PIB em relação ao número de escolas
_ = sns.barplot(x='NO_MUNICIPIO', y='PIB', data=grafico_pib, palette='Blues_d')
_.set_xticklabels(_.get_xticklabels(),rotation=45)
_.set_ylabel('PIB (R$ 1000)')
_.set_xlabel('Nome Cidade')
_.set_title('20 Cidade com Maior PIB por Quantidade de Escolas')
```

```python
#printando o PIB da cidade pelo PIB/escolas
sns.stripplot(x='PIB_Geral', y='PIB', data=merge_cnod2)
_.set_xticklabels(_.get_xticklabels(),rotation=45)
_.set_ylabel('PIB (R$ 1000)/Escolas')
_.set_xlabel('PIB (R$ 1000)')
_.set_title('Relação do PIB pelo PIB/Escolas')
```
